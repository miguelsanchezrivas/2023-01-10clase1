```python
import pandas as pd
import numpy as np

path = "/content/drive/MyDrive/2023-01-10 mineria/Challenge_4y5.csv"
```

```python
df = pd.read_csv(path)
#muestra = df.head(100)
```

```python
df["totales"] = df.iloc[:, 2:].sum(axis=1)
df.head()
```

```python
df.info()
```

```python
# #cambiar de float a int. No es necesario
# muestra["sem44"] = muestra["sem44"].astype(int)
# muestra["sem43"] = muestra["sem43"].astype(int)
# muestra["sem42"] = muestra["sem42"].astype(int)
# muestra["sem41"] = muestra["sem41"].astype(int)
```

```python
# Añadir columnas de porcentajes de cada fila

df["sem44%"] = ((df.loc[:, "sem44"]) / (df.loc[:, "totales"]))*100
df["sem43%"] = ((df.loc[:, "sem43"]) / (df.loc[:, "totales"]))*100
df["sem42%"] = ((df.loc[:, "sem42"]) / (df.loc[:, "totales"]))*100
df["sem41%"] = ((df.loc[:, "sem41"]) / (df.loc[:, "totales"]))*100
df.head()
```

```python
# Group by Producto, client is too detailed?
df_por_producto = df.groupby(["Producto"]).size()
df_por_producto = df_por_producto.to_frame()
df_por_producto
```
