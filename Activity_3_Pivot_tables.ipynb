# Activity 3: Pivot tables


## 1. Download the 'anthropometry-dataset.csv' file (if you haven't already downloaded it).

##2. Open the .csv file using python.

```python
import pandas as pd
import numpy as np

path = "/content/drive/MyDrive/2023-01-10 mineria/antropometria-dataset.csv"

antro = pd.read_csv(path)
antro
```

```python
antro.columns

# antro[["desc_ent", "sexo", "peso"]]             #esto va a mostrar las dos columnas completas
# antro
```

```python
favs = ['desc_ent', 'sexo', 'peso']
antro_favs = antro[favs]
antro_favs
```

##3. Generate a pivot table showing the average weight of people by state.


```python
pd.pivot_table(antro_favs, 
               index = ["desc_ent"],
               aggfunc = "mean", 
               margins = True       #agrega el total, en la ultima fila
               )   
```

##4. Add columns showing the average weight by gender by state.


```python
pd.pivot_table(antro_favs, 
               index = ["desc_ent"],
               aggfunc = "mean", 
               columns = "sexo",
               margins = True       #agrega el total, en la ultima fila
               )  
```

```python
antro_favs_pivot =  pd.pivot_table(antro_favs, 
                                  index = ["desc_ent"],
                                  aggfunc = "mean", 
                                  columns = "sexo",
                                  margins = True       #agrega el total, en la ultima fila
                                  )  

antro_favs_pivot.sort_values(by= "1")


#antro_favs_pivot.nlargest(5, "sexo")
```
