## boxplots by sex

```python
fig = plt.figure()

ax1 = fig.add_subplot(121)
sns.boxplot(data=hombres.loc[:, ["peso"]], 
            whis=(0, 100)) #include outliers in the line

ax2 = fig.add_subplot(122, sharey=ax1) #sharing same y axis with ax1
sns.boxplot(data=mujeres.loc[:, ["peso"]])# Female Boxplot

ax1.title.set_text('Peso de los hombres')
ax2.title.set_text('Peso de las mujeres')
plt.xticks(rotation = 90)

fig.tight_layout(pad = 0.5) #space between the figures

plt.setp(ax2.get_yticklabels(), visible=False) #to make the labels invisible on just one axis.
plt.show()

fig = plt.figure()

ax1 = fig.add_subplot(121)
sns.boxplot(data=hombres.loc[:, ["cintura"]],whis=(0, 100)) #include outliers in the line

ax2 = fig.add_subplot(122, sharey=ax1) #sharing same y axis with ax1
sns.boxplot(data=mujeres.loc[:, ["cintura"]])# Female Boxplot

ax1.title.set_text('Cintura de los hombres')
ax2.title.set_text('Cintura de las mujeres')
plt.xticks(rotation = 0)

fig.tight_layout(pad = 0.5) #space between the figures

plt.setp(ax2.get_yticklabels(), visible=False) #to make the labels invisible on just one axis.
plt.show()
```

# 4.Write the code in Python to generate a scatterplot to analyze the relationship between the variables 'weight' and 'waist'.

```python
sns.lmplot(x='peso', 
           y='cintura',
           data=df
          )
plt.title('Peso y cintura - todos')
plt.show()

sns.lmplot(x='peso', 
           y='cintura', 
           data=df,
           fit_reg=False,
           scatter_kws = { 'alpha' : 0.1 },  
           hue='sexo')   # Color by evolution stage

plt.title('peso y cintura por sexo')
plt.show()
```

### 'weight' outliers: 2,826

```python
# Yes. mean + std = 64.6 + 17.6 = 82.29
df["peso"].describe()
```

```python
# Outliers are mean + std 
peso_outliers = df["peso"].mean() + df["peso"].std()
peso_outliers
```

```python
# There are 2,826 observations > peso_outliers  
df["peso"].loc[df["peso"]>peso_outliers]
```

```python
numero_de_peso_outliers = df["peso"].loc[df["peso"]>peso_outliers]
numero_de_peso_outliers.to_frame().shape
```

### 'waist' outliers: 2,417

```python
# Yes. mean + std = 97.77 + 16.95 = 104.72
df["cintura"].describe()
```

```python
# Outliers are mean + std 
cintura_outliers = df["cintura"].mean() + df["cintura"].std()
cintura_outliers
```

```python
# There are 2,417 observations > cintura_outliers  
df["cintura"].loc[df["cintura"]>cintura_outliers]
```

```python
numero_de_cintura_outliers = df["cintura"].loc[df["cintura"]>cintura_outliers]
numero_de_cintura_outliers.to_frame().shape
```
